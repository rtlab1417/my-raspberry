## Materials

*   RPI 3B+
*   an LED (any colour)
* a resistor, 330 Ohm (use for limiting the current. Higher resistance would result in dimmer light.)

## Python 3

Code



```
import RPi.GPIO as GPIO
import time
GPIO.setmode(GPIO.BCM)
GPIO.setwarnings(False)
GPIO.setup(18,GPIO.OUT)
print("LED on")
GPIO.output(18,GPIO.HIGH)
time.sleep(1)
print("LED off")
GPIO.output(18,GPIO.LOW)
```
